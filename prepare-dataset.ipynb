```python
from scipy.io import loadmat
from datetime import datetime
import json
import os
import cv2
```

```python
def calc_age(taken, dob):
    birth = datetime.fromordinal(max(int(dob) - 366, 1))
    if birth.year > taken:
        left = birth.year
        right = taken
    else:
        left = taken
        right = birth.year
    if birth.month < 7:
        return left - right
    else:
        return left - right - 1

def get_meta(mat_path, db):
    meta = loadmat(mat_path)
    full_path = meta[db][0, 0]["full_path"][0]
    dob = meta[db][0, 0]["dob"][0]  # Matlab serial date number
    gender = meta[db][0, 0]["gender"][0]
    photo_taken = meta[db][0, 0]["photo_taken"][0]  # year
    face_score = meta[db][0, 0]["face_score"][0]
    second_face_score = meta[db][0, 0]["second_face_score"][0]
    age = [calc_age(photo_taken[i], dob[i]) for i in range(len(dob))]

    return full_path, dob, gender, photo_taken, face_score, second_face_score, age
```

```python
dataset=get_meta('wiki_crop/wiki.mat', 'wiki')
filter_dataset = []
for i in range(len(dataset[0])):
    if dataset[-1][i] < 10:
        continue
    filter_dataset.append([dataset[0][i],dataset[2][i],dataset[-1][i]])
```

```python
genders = ['female','male']
origin_path = 'wiki_crop/'
data_path = 'data/'
min_size = (20,20)
haar_scale = 1.1
min_neighbors = 3
haar_flags = 0
```

```python
def detectFace(image, faceCascade):
    image = cv2.equalizeHist(image)
    faces = faceCascade.detectMultiScale(image, haar_scale, min_neighbors, haar_flags, min_size)
    return faces

def faceCrop(imagelist, boxScale=1):
    faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
    for img, gender, age in imagelist:
        img = img[0]
        read_img = cv2.imread(origin_path+img)
        if len(read_img.shape) > 2:
            read_img = cv2.cvtColor(read_img, cv2.COLOR_RGB2GRAY)
        faces = detectFace(read_img, faceCascade)
        if len(faces)>0:
            for (x,y,w,h) in faces:
                try:
                    cropped_img = read_img[y:y+h,x:x+w]
                    fname, ext = os.path.splitext(img)
                    fname = fname.split('/')
                    cv2.imwrite(data_path+genders[int(gender)]+'--'+str(age)+'--'+fname[-1]+ext, cropped_img)
                except:
                    pass
        else:
            print('no face found', img)
```

```python
faceCrop(filter_dataset)
```
